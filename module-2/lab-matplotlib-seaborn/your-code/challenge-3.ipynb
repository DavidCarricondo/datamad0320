#### Use subplots to plot the distribution of the `Age` variable with bins equal to `10`, `20` and `50`.

```python
fig, axes = plt.subplots(1,3, figsize=(30,8))
bins = [10, 20, 30]
for i, e in enumerate(bins):
    sns.distplot(titanic.Age, bins = e, ax=axes[i]).set_title(f'Age distribution {e} bins', fontsize=30)
plt.tight_layout()
```

#### Whether the bin size affects your plot? Why?

It does affect how it looks in that the less bins, the more aggregated is the data and slightly differnt patterns may seem to show. For example, in the first plot it seems that there was a high number of people that were around 30 years old. However, in the last plot, we see that that peak was actually people in their early twenties, and there was a drop in the mid twenties. However, the density line does not change with the number of bins, because it is calculated in the same way regardless of the number of bins.

#### We suspect that there is a linear relationship between `Fare` and `Age`. Use the appropriate plot to show the relationship between these 2 features. There are 2 ways, please do it both ways.

*Hint: One of the ways involves using Seaborn*

```python
plt.scatter(titanic.Age, titanic.Fare)
plt.title('Relationship between the age and the fare')
plt.xlabel('Age')
plt.ylabel('Fare')
```

```python
scat = sns.scatterplot(titanic.Age, titanic.Fare)
scat.set_title('Relationship between the age and the fare')
```

```python
reg = sns.jointplot('Age', 'Fare', data=titanic, kind='reg')
plt.subplots_adjust(top=0.95)
reg.fig.suptitle('Relationship between the age and the fare')
```
